# Lineups

```python
teams = teams.get_teams()
mia = [x for x in teams if x['full_name'] == 'Miami Heat'][0]
mia_id = mia['id']
```

```python
heat_lineup = teamdashlineups.TeamDashLineups(
      date_from_nullable = "",
      date_to_nullable = "",
      game_id_nullable = "",
      game_segment_nullable = "",
      group_quantity = 5,
      last_n_games = 0,
      league_id_nullable = "00",
      location_nullable = "",
      measure_type_detailed_defense = "Base",
      month = 0,
      opponent_team_id = 0,
      outcome_nullable = "",
      pace_adjust = "N",
      plus_minus = "N",
      po_round_nullable = "",
      per_mode_detailed = "PerGame",
      period = 0,
      rank = "N",
      season = "2023-24",
      season_segment_nullable = "",
      season_type_all_star = "Regular Season",
      shot_clock_range_nullable = "",
      team_id = mia_id,
      vs_conference_nullable = "",
      vs_division_nullable = ""
      )
    
df = heat_lineup.get_data_frames()[1] # get the data 
```

# Game Log

```python
seasons = ['2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24']
```

```python
all_seasons_logs_df = pd.DataFrame()

for season in seasons:
    mia_logs = teamgamelogs.TeamGameLogs(team_id_nullable=mia_id, season_nullable=season)
    season_logs_df = mia_logs.get_data_frames()[0]
    #season_logs_df['SEASON'] = season  
    all_seasons_logs_df = pd.concat([all_seasons_logs_df, season_logs_df], ignore_index=True)
```

```python
all_seasons_logs_df.rename(columns={'FG3M':'3PM', 'FG3A':'3PA', 'FG3_PCT':'3P_PCT'}, inplace=True)
all_seasons_logs_df
```

```python
all_seasons_logs_df = all_seasons_logs_df.drop(['TEAM_ID', 'GAME_ID' , 'AVAILABLE_FLAG'], axis=1)
all_seasons_logs_df
```

```python
# Convert Game_Date to a datetime
all_seasons_logs_df['GAME_DATE'] = pd.to_datetime(all_seasons_logs_df['GAME_DATE'], format='mixed')
```

```python
all_seasons_logs_df.to_csv('MiamiHeatGames.csv', index=False)
```
